```python
#choose a good inital starting point near the extrema you want, increases chance of graphing it right
import matplotlib.pyplot as plt
import numpy as np

def gradient_descent(f, learning_rate, initial_point, max_steps=10000, tolerance=1e-10):
    """
    Perform gradient descent to find a local minimum of a given function.
    
    Parameters:
    f (function): The function to minimize.
    learning_rate (float): The step size.
    initial_point (float): Starting point for the descent.
    max_steps (int): Maximum number of steps to take (default=10000).
    tolerance (float): Threshold for |f'(x)| to stop iteration (default=1e-10).
    
    Returns:
    tuple: The last x_n and f(x_n), rounded to three decimal places.
    """
    def deriv(f, base_point): 
        """Estimate the derivative of f at base_point using the symmetric approximation."""
        epsilon = 1e-10
        return (f(base_point + epsilon) - f(base_point - epsilon)) / (2 * epsilon)
    
    x_coords = [initial_point]
    y_coords = [f(initial_point)]
    
    for _ in range(max_steps):
        grad = deriv(f, x_coords[-1])  # Compute derivative
        if abs(grad) < tolerance:  # Stop if the derivative is near zero
            break
        # Update point
        new_x = x_coords[-1] - learning_rate * grad
        x_coords.append(new_x)
        y_coords.append(f(new_x))
    
    # Plot the function and steps
    plot_range = np.linspace(min(x_coords) - 0.5, max(x_coords) + 0.5, 1000)
    function_range = [f(i) for i in plot_range]
    plt.plot(plot_range, function_range, label="f(x)")
    plt.plot(x_coords, y_coords, 'o-', label="Steps")
    plt.legend()
    plt.title("Gradient Descent Visualization")
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.show()
    
    return round(x_coords[-1], 3), round(y_coords[-1], 3)

# Example Functions

# Test case: f(x) = x^2
def f1(x):
    return x**2

# Test case: f(x) = x^4 - 2x^2
def f2(x):
    return x**4 - 2*x**2

# Symmetrized |x| function
def funny_function(x):
    if x > 0:
        return x**x
    elif x == 0:
        return 1
    else:
        return abs(x)**abs(x)

# Gradient descent to find local minima for |x|
def abs_function(x):
    return abs(x)

# Run examples
print("Testing f(x) = x^2")
print(gradient_descent(f1, 0.2, 5))

print("Testing f(x) = x^4 - 2x^2")
print(gradient_descent(f2, 0.1, -1))  # One local minimum
print(gradient_descent(f2, 0.1, 1))   # Another local minimum

print("Testing symmetrized x^x function")
print(gradient_descent(funny_function, 0.01, 0.1))  # Careful with learning rate!

print("Testing f(x) = |x|")
print(gradient_descent(abs_function, 0.2, 5))
```

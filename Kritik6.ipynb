```python
import numpy as np
import matplotlib.pyplot as plt

# Define the normal density function
def normal_density(mean, variance, x):
    sigma = np.sqrt(variance)  # Standard deviation
    return (1 / np.sqrt(2 * np.pi * variance)) * np.exp(-((x - mean) ** 2) / (2 * variance))

# Numerical integration using the Trapezoidal Rule
def integration(mean, variance, a, b, num_points=1000):
    x_vals = np.linspace(a, b, num_points)  # Generate points between a and b
    y_vals = normal_density(mean, variance, x_vals)  # Evaluate density function at each point
    dx = (b - a) / (num_points - 1)  # Step size
    return np.trapz(y_vals, dx=dx)  # Approximate integral using trapezoidal rule

# Plot the normal density function for male height distribution
def plot_normal_density():
    x_vals = np.linspace(150, 200, 1000)  # X range from 150 cm to 200 cm
    μ, σ2 = 171, 7.1**2  # Mean and variance of male height distribution
    
    plt.plot(x_vals, normal_density(μ, σ2, x_vals), label=f"μ={μ}, σ²={σ2}", color="blue")
    plt.xlabel("Height (cm)")
    plt.ylabel("Density")
    plt.title("Normal Distribution of Male Heights")
    plt.legend()
    plt.grid()
    plt.show()

# Compute probability of a randomly chosen male being between 162 cm and 190 cm
μ, σ2 = 171, 7.1**2
prob = integration(μ, σ2, 162, 190)
print(f"Probability that a randomly chosen male has a height between 162 cm and 190 cm: {prob:.4f}")

# Plot the normal distribution
plot_normal_density()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# (a) Uniform Distribution Expected Value
def uniform_expected_value(a, b):
    return (a + b) / 2

# (b) Exponential Distribution Expected Value
def exponential_expected_value(lmbda):
    return 1 / lmbda

# (c) Normal Distribution Density Function
def normal_density(mean, variance, x):
    sigma = np.sqrt(variance)
    return (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-((x - mean) ** 2) / (2 * variance))

# Numerical Integration Using Trapezoidal Rule
def numerical_integration(func, a, b, n=1000):
    x = np.linspace(a, b, n)
    y = func(x)
    dx = (b - a) / (n - 1)
    return np.sum((y[:-1] + y[1:]) * dx / 2)  # Trapezoidal sum

# Dosage Function
def dosage_function(x):
    return 2.38 * x**2

# Expected Dosage Calculation using Numerical Integration
def expected_dosage(mean, variance, a, b, n=1000):
    integrand = lambda x: dosage_function(x) * normal_density(mean, variance, x)
    return numerical_integration(integrand, a, b, n)

# Given values
a, b = 0, 10  # Example values for uniform distribution
lambda_pandemic = 1 / 50
mean_height = 171
variance_height = 7.1**2

# Compute results
exp_value_uniform = uniform_expected_value(a, b)
exp_value_pandemic = exponential_expected_value(lambda_pandemic)
exp_dosage = expected_dosage(mean_height, variance_height, 150, 200)  # Integrate over reasonable height range

# Display results
print(f"(a) Expected value of Uniform[{a}, {b}]: {exp_value_uniform:.2f}")
print(f"(b) Expected time between pandemics: {exp_value_pandemic:.2f} years")
print(f"(c) Estimated average drug dosage: {exp_dosage:.2f}")

# Plot normal density function
x_vals = np.linspace(150, 200, 1000)
y_vals = normal_density(mean_height, variance_height, x_vals)
plt.plot(x_vals, y_vals, label="Normal Distribution PDF")
plt.xlabel("Height (cm)")
plt.ylabel("Density")
plt.title("Normal Distribution of Male Height")
plt.legend()
plt.show()
```
